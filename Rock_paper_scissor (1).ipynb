```python
# Rock Paper Scissor AI game
# Press 's' to start/replay the game
# Press 'r' to restart the game from score (0,0)
# Press 'q' to quit the game

import cv2
import cvzone
from cvzone.HandTrackingModule import HandDetector
import time
import random

cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

detector = HandDetector(maxHands=1)


timer = 0
startGame = 0
Result = False
Scores = [0,0] # [Computer , Player]



while True:
    imgBkg = cv2.imread("E:/RPS/Res/background.png")
    success, img = cap.read()
    
    imgresize = cv2.resize(img,(0,0),None,0.875,0.875)
    imgScale = imgresize[: ,80:480]   #To crop from center and not from the sides
    
    hands, img = detector.findHands(imgScale)
    
    if startGame:
        
        if Result is False:
            timer = time.time()-initialTime
            cv2.putText(imgBkg,str(int(timer)),(605,435),cv2.FONT_HERSHEY_PLAIN, 6,(255,0,255),4) #Displays the timer
            
            if timer>3:
                Result = True
                timer = 0    
                
                if hands:
                    PlyrMove = None
                    hand = hands[0]
                    fingers = detector.fingersUp(hand) #To detect number of fingers
                    
                    if fingers == [0,0,0,0,0]: # Hand position for rock
                        PlyrMove = 1
                    if fingers == [1,1,1,1,1]: # Hand position for paper
                        PlyrMove = 2
                    if fingers == [0,1,1,0,0]: # Hand position for scissor
                        PlyrMove = 3
                    
                    randNumber = random.randint(1,3)
                    imgAI = cv2.imread(f"E:/RPS/Res/{randNumber}.png",cv2.IMREAD_UNCHANGED)
                    imgBkg = cvzone.overlayPNG(imgBkg,imgAI,(165,288))
                    
                    # When the player wins
                    if (PlyrMove == 1 and randNumber == 3) or \
                        (PlyrMove == 2 and randNumber == 1) or \
                        (PlyrMove == 3 and randNumber == 2):
                        Scores[1] += 1
                        
                     # When the Computer wins
                    if (PlyrMove == 3 and randNumber == 1) or \
                        (PlyrMove == 1 and randNumber == 2) or \
                        (PlyrMove == 2 and randNumber == 3):
                        Scores[0] += 1

    imgBkg[231:651,805:1205] = imgScale
    
    if Result:
        imgBkg = cvzone.overlayPNG(imgBkg,imgAI,(165,288))
        
    cv2.putText(imgBkg,str(Scores[0]),(410,215),cv2.FONT_HERSHEY_PLAIN, 4,(255,255,255),6)
    cv2.putText(imgBkg,str(Scores[1]),(1112,215),cv2.FONT_HERSHEY_PLAIN, 4,(255,255,255),6)
  
    cv2.imshow("Back_ground",imgBkg)
    
    key = cv2.waitKey(1)

    if key == ord('s'):
        startGame = True
        initialTime = time.time()
        Result = False
        
    if key == ord('r'):
        Scores = [0,0]
        startGame = True
        initialTime = time.time()
        Result = False
        
    if key == ord('q'):
        break;
        
cap.release()
cv2.destroyAllWindows()
    
        
        
```
